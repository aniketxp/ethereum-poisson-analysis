```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Simulating Ethereum transaction arrivals (Poisson Distribution)
lam = 10  # Average transactions per minute
num_minutes = 1000  # Simulating 1000 minutes
tx_counts = np.random.poisson(lam, num_minutes)

# Plot histogram
plt.figure(figsize=(10, 5))
plt.hist(tx_counts, bins=30, alpha=0.7, color="purple", edgecolor="black")
plt.axvline(lam, color="red", linestyle="dashed", linewidth=2, label=f"Mean: {lam}")
plt.xlabel("Transactions per Minute")
plt.ylabel("Frequency")
plt.title("Ethereum Transaction Distribution (Poisson)")
plt.legend()
plt.grid()
plt.show()
```
